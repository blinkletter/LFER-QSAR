# Data Handling Examples with Pandas
Let us try some data manipulation of a table. Imagine we have a table of Hammett parameters. Let us load in the data file and start exploring.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

df = pd.read_csv("LFER_williams_acidity.csv", delimiter = ",", index_col="Substituent", comment = "#") # The fields are separated by commas.
display(df)
```

```python
np.isnan(df.at["p-CHO","BzOH_pKa"])
print(df.loc["p-CHO"])
print()
print(df.loc["p-CHO"]["BzOH_pKa"])
print()
print(df.loc[["p-CHO","p-CH3"]][["BzOH_pKa","Dimethylaniline_pKa"]])
print()
```

```python
df = pd.read_csv("LFER_gas-aqueous_acidity.csv", delimiter = ",", index_col="X") # The fields are separated by commas.
#df = pd.read_csv("LFER_gas-aqueous_acidity.csv") 
#display(df)
df.loc[np.isnan(df[:]["s+"]), ["s+"]]=df.loc[np.isnan(df[:]["s+"]), ["Sigma"]]
df.loc[np.isnan(df[:]["s+"])]
a = df.loc[np.isnan(df[:]["s+"]), ["Sigma"]]
a
df.loc[np.isnan(df[:]["s+"]), ["s+"]] = df.loc[np.isnan(df[:]["s+"]), ["Sigma"]]
display(df.loc[np.isnan(df[:]["s+"]), ["s+"]])
display(df)
df.loc[np.isnan(df[:]["s+"]), ["s+","Sigma"]]
```

```python
df = pd.read_csv("LFER_gas-aqueous_acidity.csv", delimiter = ",", index_col="X") # The fields are separated by commas.
#df = pd.read_csv("LFER_gas-aqueous_acidity.csv") 
x, y = df.BzOH, df.PhOH

#plt.plot(df["BzOH"], df["PhOH"], "ko")
fig, ax = plt.subplots()
ax.scatter(x, y,      
           s=50,                  # Market size, can be an array
           marker = "o",
           facecolor = "white",
           edgecolor = "black",
           linewidths = 1,
           alpha = 1
          )
#'''
for i, txt in enumerate(df.index):
    if txt[0] == "p":
        offset = (-20,10)
        position = ("bottom","right")
    elif txt[0] == "m":
        offset = (20,-10)
        position = ("top","left")
    else:
        offset = (-20,0)
        position = ("bottom","right")
           
    ax.annotate(txt, (df.BzOH[i], df.PhOH[i]), 
                xycoords = "data",
                va = position[0],
                ha = position[1],
                arrowprops = dict(
                                  arrowstyle = "->",
                                  ),
                textcoords = "offset points",
                xytext = offset
               )
#'''

ax.set_ylim(-3,25)               

ax.set_xlabel("BzOH")               
ax.set_ylabel("PhOH")
ax.set_title("Comparing Free Energy of Gas Phase Acidities")          

plt.show()
```
