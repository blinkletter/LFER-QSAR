# Gas Phase Free Energies of Proton Transfer
The free energy difference, $\Delta G^{\circ}$, for transfer of a proton between the substituted and unsubstituted form of an acid can be measured. Let us see if the gas phase acidity of benzoic acids correlate to phenols. And then we will see if either of these correlate to Hammett $\sigma$ values.

We will begin the usual way by loading the modules, defining functions and then importing data from files into dataframes.  We will fill the $\sigma$ values in the table of Hammett parameters across into empty $\sigma^+_p$ and $\sigma_p^-$ slots as usual. beware this may cause errors if there is a $\sigma^+_p$ or $\sigma_p^-$ missing from the table.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

def fill_sigma(df):     # a function to fill in sigma for empty spaces in s+ and S-
    for z in df.index:
        if np.isnan(df.loc[z,"s_plus"]):
            df.loc[z,"s_plus"] = df.loc[z,"sigma"]
        if np.isnan(df["s_minus"][z]):
            df.loc[z,"s_minus"] = df.loc[z,"sigma"]
    return(df)


df1 = pd.read_csv("../data/LFER_Williams.csv", 
                  delimiter = ",", 
                  index_col="Substituent", 
                  comment = "#", 
                  skipinitialspace = True
                 ) 
df2 = pd.read_csv("../data/ACID_gas_acidity.csv", 
                  delimiter = ",", 
                  index_col="Substituent", 
                  comment = "#", 
                  skipinitialspace = True
                 ) 
df1 = fill_sigma(df1)

df3 = df1.join(df2)
#display(df3.head(5))

data = df3[["sigma", "s_plus", "s_minus", "BzOH", "PhOH"]]
data = data.dropna()
#display(data)
```

## Investigating The Data
Let us make a **quick plot** of the two sets of $\Delta G^{\circ}$. remember that $\log(K_a) \propto \Delta G^{\circ}$ so this will be just like a log-log plot.

```python
x = data["BzOH"]
y = data["PhOH"]

plt.plot(x, y, "ko")

plt.title("Comparing $\Delta G^{\circ}$ in gas phase acidity")               
plt.xlabel(r"$\Delta G^{\circ}$ BzOH")               
plt.ylabel(r"$\Delta G^{\circ}$ PhOH")   

#plt.xlim([-.4,1.3])
#plt.ylim([-10.5,-7])

plt.show()

```

# Analyze the Data
That didn't look so good. If we look through the data we see that the **m-substituted** pairs seem to fall in a line while the **p-substituted** points scatted above that line. We can visualize this by **separating** the two sets. How could we do that? There are many ways,

We will **sort** the data by the index entries. All the "m"s will end up together. then we just count down and pick them into the two groups.

Again, we see that the m-substituted groups line up but many of the p-substituted do not. In the more electron-withdrawing substitient cases the effect on phenols seems to be greater than the corresponding effect on benzoic acids.

```python
sort = data.sort_index()
#display(sort)
meta = sort[0:10]
para = sort [10:]
#display(meta)
#display(para)

x, y = meta["BzOH"], meta["PhOH"]
plt.plot(x, y, "ko")

x, y = para["BzOH"], para["PhOH"]
plt.plot(x, y, "kx")

#y_list = 2.07*x-9.82
#plt.plot(x, y_list, "k-")

plt.title("Comparing $\Delta G^{\circ}$ in gas phase acidity")               
plt.xlabel(r"$\Delta G^{\circ}$ BzOH")               
plt.ylabel(r"$\Delta G^{\circ}$ PhOH")   
#plt.xlim([-.4,1.3])
#plt.ylim([-10.5,-7])

plt.show()
```

## Again, with $\sigma$

Here we will plot the benzoic acid and the phenol $\Delta G^{\circ}$ values against $\sigma$. These $\sigma$ are defined by the $\Delta G^{\circ}$ for acid dissociation of benzoic acid in water. So I hope to see good correlation with gas phase acidities for the benzoic acid. Let us see...

You will see that we dropped the data for $-NH_2$ substitued acids. this is an additional acid group in the mix and so I wanted to avoid complications. It also didn't fit the line. Is that justification enough? Not!



```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(12,4))


x, y = data["sigma"], data["BzOH"] 
to_drop = ["p-NH2", "m-NH2"]
x.drop(labels=to_drop, inplace=True)               # deletion of data points
y.drop(labels=to_drop, inplace=True)


ax[0].plot(x,y, "ko")            # plot the data from the dataframe
ax[0].set(ylabel=r"$\Delta G^{\circ}$ BzOH", xlabel=r"$\sigma$")      

linefit = stats.linregress(x,y)                  # perform a line fit
calc_y = linefit.intercept + linefit.slope * x   # calculate the y values for the line fit 
ax[0].plot(x, calc_y, "b-")  # plot the line fit


x, y = data["sigma"], data["PhOH"]  
to_drop = ["p-NH2", "m-NH2"]
x.drop(labels=to_drop, inplace=True)               # deletion of data points

y.drop(labels=to_drop, inplace=True)

ax[1].plot(x,y, "ko")            # plot the data from the dataframe
ax[1].set(ylabel="$\Delta G^{\circ}$ PhOH", xlabel=r"$\sigma$")   

#linefit = stats.linregress(x,y)                  # perform a line fit
#calc_y = linefit.intercept + linefit.slope * x   # calculate the y values for the line fit 
calc_y = 0.29 + 19.87 * x   # using values for the m-substituents 
ax[1].plot(x, calc_y, "b-")  # plot the line fit

#print("slope = {:.2f} +/- {:.2f}".format(linefit.slope, 2*linefit.stderr))
#print("intercept = {:.2f} +/- {:.2f}".format(linefit.intercept, 2*linefit.intercept_stderr))

plt.show()
```

We have learned that electron accepting substituents can have an outsized effect when in the **para position** of a phenol. So we should use Hammett parameters that take this into account.  The **Brown-Okamoto** $\sigma_p^-$ parameters may give better correlations. What do you think?



```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(12,4))



x, y = data["s_minus"], data["BzOH"] 
to_drop = ["p-NH2", "m-NH2"]
x.drop(labels=to_drop, inplace=True)               # deletion of data points
y.drop(labels=to_drop, inplace=True)

ax[0].plot(x,y, "ko")            # plot the data from the dataframe
ax[0].set(ylabel=r"$\Delta G^{\circ}$ BzOH", xlabel=r"$\sigma^-$") 

linefit = stats.linregress(x,y)                  # perform a line fit
calc_y = linefit.intercept + linefit.slope * x   # calculate the y values for the line fit 
ax[0].plot(x, calc_y, "b-")  # plot the line fit




x, y = data["s_minus"], data["PhOH"]  
to_drop = ["p-NH2", "m-NH2"]
x.drop(labels=to_drop, inplace=True)               # deletion of data points
y.drop(labels=to_drop, inplace=True)

ax[1].plot(x,y, "ko")            # plot the data from the dataframe
ax[1].set(ylabel="$\Delta G^{\circ}$ PhOH", xlabel=r"$\sigma^-$")       

linefit = stats.linregress(x,y)                  # perform a line fit
calc_y = linefit.intercept + linefit.slope * x   # calculate the y values for the line fit 
#calc_y = 0.29 + 19.87 * x   # using values for the m-substituents 

ax[1].plot(x, calc_y, "b-")  # plot the line fit

plt.show()
```
