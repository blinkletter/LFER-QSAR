# pKa Values from R. Williams
This table presents selected series of $pK_a$ values from the table compiled by Jencks and Williams. this table is available from the ASmerican Chemical Society at https://organicchemistrydata.org/hansreich/resources/pka/

 The data series are as follows:
 
- "Substituent" is the position and structure of the substituent
- "BzOH pKa" is acidity of substituted BENZOIC ACIDS in water
- "PhOH pKa" is acidity of substituted PHENOLS in water
- "Aniline pKa" is acidity of substituted ANILINESS in water
- "Dimethylaniline pKa" is acidity of substituted N,N-DIMETHYLANILINES in water
- "Pyridine pKa" is acidity of substituted PYRIDINES in water

 

```python
import pandas as pd

# The fields are separated by commas. Comments are enabled
df = pd.read_csv("data/LFER_williams_acidity.csv", delimiter = ",", index_col="Substituent", comment = "#") 
display(df)
```
